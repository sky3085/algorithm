```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd
import copy
import sys, os

#https://skkim1080.tistory.com/entry/Python-%ED%8C%8C%EC%9D%B4%EC%8D%AC-%EC%8B%A4%ED%96%89%ED%8C%8C%EC%9D%BC%EC%97%90-chromedriver-%EC%B6%94%EA%B0%80%ED%95%98%EA%B8%B0%ED%8F%AC%ED%95%A8%EC%8B%9C%ED%82%A4%EA%B8%B0
#pyinstaller --add-binary "chromedriver.exe;." .\testexe.py

input_name=input('코트를 입력해주세요 ex(신도림 5월 주말 A코트 주간) :')
id=input('아이디를 입력해주세요 :')
password=input('비밀번호 입력해주세요 :')
date_input=input('원하는 날짜ex(20220429) :')
date=f'cal_{date_input}'
time_pick=input('원하는 예약 시간 ex(11시) :')
program_rum=input('예약 열리는 시간ex(7:0:0 , 7:2:3) :')

#webdriver_options = webdriver.ChromeOptions()
#webdriver_options.add_argument('headless')

URL = 'https://yeyak.seoul.go.kr/web/search/selectPageListDetailSearchImg.do?code=T100&dCode=T108'
if  getattr(sys, 'frozen', False): 
    chromedriver_path = os.path.join(sys._MEIPASS, "chromedriver.exe")
    driver = webdriver.Chrome(chromedriver_path)
else:
    driver = webdriver.Chrome()
driver.implicitly_wait(10)
driver.maximize_window()
driver.get(URL)#신도림 4월 주말 A코트 주간

#로그인
driver.find_element_by_xpath('//*[@id="header"]/div[1]/div/div[1]/a').click()
driver.find_element_by_xpath('//*[@id="userid"]').send_keys(id)
driver.find_element_by_xpath('//*[@id="userpwd"]').send_keys(password)
driver.find_element_by_xpath('//*[@id="addUserForm"]/div/button').click()
driver.implicitly_wait(10)
driver.find_element_by_xpath('//*[@id="search2"]/fieldset/div[2]/input').send_keys(input_name)
driver.find_element_by_xpath('//*[@id="search2"]/fieldset/button[1]').click()
now = time.localtime()
ck = str(now.tm_hour)+ ':' + str(now.tm_min) + ':' + str(now.tm_sec)

while True :
    if ck!=program_rum:
        driver.implicitly_wait(10)
        now = time.localtime()
        ck = str(now.tm_hour)+ ':' + str(now.tm_min) + ':' + str(now.tm_sec)
        print(ck)
    elif(ck==program_rum):
        print('실행')
        #첫번째 인풋으로 검색한거 클릭
        driver.find_element_by_xpath('//*[@id="contents"]/div[3]/ul/li/a').click()
        driver.find_element_by_xpath('//*[@id="contents"]/div[1]/div/div[2]/button').click()#팝업 닫기
        driver.implicitly_wait(10)

        #만약에 7시 정각에 날짜가 "예약가능"이 아닐때 새로고침
        while driver.find_element_by_xpath(f'//*[@id="{date}"]').get_attribute('title') == '예약불가':
            time.sleep(0.7)
            #새로고침
            driver.refresh()
            driver.find_element_by_xpath('//*[@id="contents"]/div[1]/div/div[2]/button').click()
            zz=driver.find_element_by_xpath(f'//*[@id="{date}"]').get_attribute('title')
            print(zz+" 새로고침하겠습니다.")
            if zz=='예약가능':
                break
        
        #날짜선택
        driver.find_element_by_xpath(f'//*[@id="{date}"]').click()
        driver.find_element_by_xpath('//*[@id="aform"]/div[1]/div[2]/div/div/a[1]').click()

        #시간 선택
        if(time_pick=='9시' or time_pick=='21시'):
            driver.find_element_by_xpath('//*[@id="unit1"]/a').click()
        elif(time_pick=='11시'):
            driver.find_element_by_xpath('//*[@id="unit2"]/a').click()
        elif(time_pick=='7시' or time_pick=='19시'):
            driver.find_element_by_xpath('//*[@id="unit0"]/a').click()
        elif(time_pick=='13시'):
            driver.find_element_by_xpath('//*[@id="unit3"]/a').click()
        elif(time_pick=='15시'):
            driver.find_element_by_xpath('//*[@id="unit4"]/a').click()
        elif(time_pick=='17시'):
            driver.find_element_by_xpath('//*[@id="unit5"]/a').click()

        driver.find_element_by_xpath('//*[@id="user_cnt_area"]/div[2]/div[1]/div/div/div/button[2]').click()
        driver.find_element_by_xpath('//*[@id="aform"]/div[3]/div[2]/div[5]/h5/span/label').click()

        #스크롤 다운
        for i in range(0,8):
            driver.find_element_by_xpath('//*[@id="aform"]/div[3]/div[2]/div[6]/div[1]/div[1]').send_keys(Keys.PAGE_DOWN)
        for i in range(0,5):
            driver.find_element_by_xpath('//*[@id="aform"]/div[3]/div[2]/div[6]/div[2]/div[1]').send_keys(Keys.PAGE_DOWN)
        for i in range(0,3):
            driver.find_element_by_xpath('//*[@id="chk_agree3_1_div"]').send_keys(Keys.PAGE_DOWN)

        driver.find_element_by_xpath('//*[@id="aform"]/div[3]/div[2]/div[6]/h5/span/label').click()
        driver.find_element_by_xpath('//*[@id="aform"]/div[3]/div[3]/div/div[3]/button').click()
        alert=driver.switch_to_alert()
        alert.accept()
        break
```
