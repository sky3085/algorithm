```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import copy
import sys, os
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
#pyinstaller --add-binary "chromedriver.exe;." .\test_s2.py

user_id=input('아이디를 입력해주세요 : ')
password=input('비밀번호 입력해주세요 : ')
year_input=input('년도 입력(2022) : ')
year=f'{year_input}년'
mon_input=input('월을 입력하세요(1, 10) : ')
mon=f'{mon_input}월'
date_input=input('날짜ex(1,3,10,11) : ')
date_input=(int)(date_input)+1
date=f'#td{date_input}'
print(date)
time_pick=input('시간 ex(08, 10, 12) : ')
program_rum=input('예약 열리는 시간ex(7:0:0 , 7:2:3) : ')


url = 'https://reserv.bucheon.go.kr/site/main/lending/lendingList?lending_inst_nm=tennis&inst_cate=01'
if  getattr(sys, 'frozen', False): 
    chromedriver_path = os.path.join(sys._MEIPASS, "chromedriver.exe")
    driver = webdriver.Chrome(chromedriver_path)
else:
    driver = webdriver.Chrome()
driver.maximize_window()
driver.get(url)

#로그인 
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="header"]/div[1]/div/div[4]/ul/li[1]/a').click()#이동
time.sleep(1)
driver.find_element(By.XPATH, '/html/body/div/div[1]/div[2]/ul/li[1]/div/a').click()#이동
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="login_id"]').send_keys(user_id)
driver.find_element(By.XPATH, '//*[@id="login_pwd"]').send_keys(password)
driver.find_element(By.XPATH, '//*[@id="memberlogin"]/div/div/div/p[1]/a/img').click()#이동

#현재 시간
now = time.localtime()
ck = str(now.tm_hour)+ ':' + str(now.tm_min) + ':' + str(now.tm_sec)

time.sleep(1.5)
print('실행')
#                                      
lis = driver.find_elements(By.XPATH, '//*[@id="CtznLendingInfoVO"]/ul[3]/li/a/span[2]')
for i in lis :
    if(i.text == '부천실내테니스장'):
        i.click()
        break

driver.find_element(By.XPATH, '//*[@id="tab2"]').click()#이동

# X월 선택     주의 년도 선택 없음
dropDownYear = Select(driver.find_element(By.XPATH, '//*[@id="d1"]'))
dropDownYear.select_by_visible_text(year)
dropDownMonth = Select(driver.find_element(By.XPATH, '//*[@id="d2"]'))
dropDownMonth.select_by_visible_text(mon)
driver.find_element(By.XPATH, '//*[@id="CtznLendingInfoVOForSearch"]/button').click()
# 날짜 시간 보고 선택
temp_date=driver.find_element(By.CSS_SELECTOR, date)
times = temp_date.find_elements(By.CSS_SELECTOR, '.tm > a')

time.sleep(1)
for i in times :
    a = i.text[:2]
    if(a == time_pick):
        i.click()
        break

driver.find_element(By.CSS_SELECTOR, '.labeldioest > input').click()
driver.find_element(By.CSS_SELECTOR, 'input#a1').click()
driver.find_element(By.CSS_SELECTOR, 'input#a2').click()
driver.find_element(By.XPATH, '//*[@id="CtznLendingBooking"]/div/a[1]').click()

time.sleep(1)
alert=Alert(driver)
alert.accept()

time.sleep(1)
alert=Alert(driver)
alert.accept()
# while True :
#     if ck!=program_rum:
#         time.sleep(0.3)
#         now = time.localtime()
#         ck = str(now.tm_hour)+ ':' + str(now.tm_min) + ':' + str(now.tm_sec)
#         print(ck)
#     elif(ck==program_rum):
#         print('실행')
#         break

```
