```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd
import copy


url = 'https://www.megabox.co.kr/booking/timetable'
driver = webdriver.Chrome('chromedriver')
driver.get(url)
driver.implicitly_wait(10)
driver.find_element_by_xpath('//*[@id="contents"]/div/div/div[1]/div[1]/ul/li[2]/a').click()
places=driver.find_elements_by_xpath('//*[@id="masterBrch"]/div/div[1]/ul/li')

cnt = 4
answer=[]
#pd.DataFrame(columns=['place', 'name', 'date', 'movie_age', 'title', 'room', 'seat_total', 'theater_type', 'time'])

#서울 경기,등 장소
for place in places:
    line={}
    place.click()
    line['place']=place.text.split('(')[0]#장소 저장
    
    #//*[@id="mCSB_4_container"]/ul/li
    #//*[@id="mCSB_5_container"]/ul/li
    names = driver.find_elements_by_xpath(f'//*[@id="mCSB_{cnt}_container"]/ul/li/button')
    
    for name in names:
        driver.implicitly_wait(10)
        if(name.is_enabled):
            name.click()
            line['name']=name.text#영화관 이름
        # 날짜
        for i in range(2,4):
            driver.implicitly_wait(10)
            a = driver.find_element_by_xpath(f'//*[@id="contents"]/div/div/div[4]/div/div[1]/div[2]/div/button[{i}]')
            if a.get_attribute('class')=='disabled':
                continue
            else:
                line['date']=a.get_attribute('date-data')
                a.send_keys(Keys.ENTER)
                
            #print(line)
            moive_list=len(driver.find_elements_by_xpath('//*[@id="contents"]/div/div/div[6]/div'))
            for b in range(1,moive_list+1):# 를 이용해서 range 반복문사용하여 xpath로 가져와 보자
                driver.implicitly_wait(10)
                movie = driver.find_element_by_xpath(f'//*[@id="contents"]/div/div/div[6]/div[{b}]')
                if(movie.get_attribute('class')=='theater-list'):
                    line['movie_age']=movie.find_element_by_css_selector('div.theater-tit > p').get_attribute('class').split('-')[-1]
                    line['title']=movie.find_element_by_css_selector('div.theater-tit a').text
                    #print(line)
                    rooms = movie.find_elements_by_css_selector('div.theater-type-box')
                    for c in range(0,len(rooms)):
                        line['room']=rooms[c].text.split('\n')[0]
                        line['seat_total']=rooms[c].text.split('\n')[1]
                        line['theater_type']=rooms[c].text.split('\n')[2]
                        #times=rooms[c].text.split('\n')[3:]
                        times=rooms[c].find_elements_by_css_selector('p.time')
                        for d in range(0,len(times)):
                            #time.sleep(1)
                            line['time']=times[d].text
                            #line['rest_seat']=times[d+1]
                            answer.append(copy.deepcopy(line))
                            
                else:
                    continue
    cnt+=1
```
